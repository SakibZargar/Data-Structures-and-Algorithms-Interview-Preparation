**1.1 Implement an algorithm to determine if a string has all unique characters. What if you cannot use additional data structures?**

(a) Use of additional data structures allowed:--------------------------------------------------------------------------------------------------------------------------

```python
def str_uniqueness_check_1(input_str):
  return (len(input_str) == len(set(input_str)))
```

```python
def str_uniqueness_check_2(input_str, input_str_type):

  if input_str_type == 'ASCII':
    no_unique_chars= 2**8
  elif input_str_type == 'Unicode':
    no_unique_chars= 2**16
  else:
    return (print('Input string type not recognized'))

  if len(input_str) > no_unique_chars:
    return (False)
  
  bool_arr= [False]*no_unique_chars
  for char in input_str:
    index= ord(char)   # the ord () function accepts a string of unit length as an argument and returns the Unicode equivalence of the passed argument
    if bool_arr[index] == True:
      return (False)
    else:
      bool_arr[index]= True

  return (True)
```

(b) Use of additional data structures NOT  allowed:-----------------------------------------------------------------------------------------------------------------

```python
def str_uniqueness_check_3(input_str):

  checker= 0
  for char in input_str:
    if (checker&(1<<(ord(char)-1))) > 0:
      return (False)
    else:
      checker= checker|(1<<(ord(char)-1))
  
  return (True)
```

**1.2 Implement a function which reverses a null-terminated string.**

```python
def reverse_string(input_str):
  
  str_reversed= ''
  i= len(input_str)-1
  while i >= 0:
    str_reversed= str_reversed + input_str[i]
    i-= 1

  return (str_reversed)
```

**1.3 Given two strings, write a method to decide if one is a permutation of the other (Anagram check).**

```python
# If spaces do not matter, then, during pre-processing remove the spaces from both the strings i.e., input_str= input_str.replace(' ','')
# If the anagram comparison is NOT case sensitive, then change the case of both the strings to upper/lower i.e., input_str= input_str.lower()
```

```python
def is_permutation_1(input_str1,input_str2):
  
  if len(input_str1) != len(input_str2):
    return (False)
  
  return (sorted(input_str1) == sorted(input_str2))
```

```python
def is_permutation_2(input_str1,input_str2,input_str_type):
  
  if len(input_str1) != len(input_str2):
    return (False)
  
  if input_str_type == 'ASCII':
    no_unique_chars= 2**8
  elif input_str_type == 'Unicode':
    no_unique_chars= 2**16
  else:
    return (print('Input string type not recognized'))

  checker= [0]*no_unique_chars

  for char in input_str1:
    checker[ord(char)]+= 1

  for char in input_str2:
    checker[ord(char)]-= 1

  return (checker == [0]*no_unique_chars)
```

**1.5 Implement a method to perform basic string compression using the counts of repeated characters (e.g., the string aabcccccaaa should become a2b1c5a3). If the compressed string would not become smaller than the original string, your method should return the original string. (Run-length Encoding)**

```python
def string_compression(input_str):
  
  if len(input_str) == 0:
    return (input_str)
  
  if len(input_str) == 1:
    return (input_str)
  
  output= ''
  count = 1
  for i in range(len(input_str)-1):
    if input_str[i] == input_str[i+1]:
      count+=1
    else:
      output= output + input_str[i] + str(count)
      count = 1
  output= output + input_str[i] + str(count)

  if len(output) > len(input_str):
    return (input_str)

  return (output)
```

**1.6 Given an image represented by an NxN matrix, where each pixel in the image is 4 bytes, write a method to rotate the image by 90 degrees. Can you do this inplace?**

```python
def rotate_matrix_inplace_90(input_mat):
  
  for i in range(len(input_mat)):
    for j in range(i,len(input_mat[0])):
      input_mat[i][j],input_mat[j][i]= input_mat[j][i],input_mat[i][j]
  
  for i in range(len(input_mat)):
    for j in range(len(input_mat[0])//2):
      input_mat[i][j],input_mat[i][-j-1]= input_mat[i][-j-1],input_mat[i][j]
  return (input_mat)
```

**1.7 Write an algorithm such that if an element in an MxN matrix is 0, its entire row and column are set to 0. Can you do this inplace?**

```python
def set_to_zero(input_mat):
  row_set= set()
  col_set= set()

  for i in range(len(input_mat)):
    for j in range(len(input_mat[0])):
      if input_mat[i][j] == 0:
        row_set.add(i)
        col_set.add(j)
  
  for i in range(len(input_mat)):
    for j in range(len(input_mat[0])):
      if i in row_set or j in col_set:
        input_mat[i][j]=0
  
  return (input_mat)
```

**1.9 Sentence reversal i.e., given a string of words, reverse all the words (e.g., 'This is the best' should become 'best the is This')**

```python
def sentence_reversal_1(input_str):
  return (' '.join(reversed(input_str.strip().split())))
```

```python
def sentence_reversal_2(input_str):
  return (' '.join(input_str.strip().split()[::-1]))
```

```python
def sentence_reversal_3(input_str):
  
  input_str= input_str.strip()

  spaces_ind= [-1]
  for i in range(len(input_str)):
    if input_str[i] == ' ':
      spaces_ind.append(i)
  spaces_ind.append(len(input_str))

  l= [0]*(len(spaces_ind)-1)
  for i in range(len(spaces_ind)-1):
    l[-i-1]= input_str[(spaces_ind[i]+1):spaces_ind[i+1]]

  return (' '.join(l))
```

**1.10 Given an array of integers (positive and negative), find the largest continuous sum and the indices of the subarray that produces that sum.**

```python
def largest_continuous_sum(input_arr):

  if len(input_arr) == 0:
    return ([0,0,0])
  if len(input_arr) == 1:
    return ([input_arr[0],0,0])  
  
  running_sum= max_sum = 0
  s= start_ind= end_ind= 0
  for i in range(len(input_arr)):
    if (running_sum+input_arr[i]) > 0:
      running_sum+= input_arr[i]
    else:
      running_sum= 0
      s= i+1                # candidate start_ind
    if max_sum < running_sum:
      max_sum= running_sum 
      end_ind= i            # the last time max_sum gets updated
      if s <= end_ind:
        start_ind= s        # the start_ind has to be less than the end_ind
    
  return ([max_sum,start_ind,end_ind])
```

**1.11 Array pair sum: Given an integer array, output the unique pairs that sum up to the specific value k (e.g., pair_sum([1,3,2,2],4) should return (1,3) and (2,2)).**

```python
def unique_pairs_sum_1(input_arr,k):
  
  output= set()
  for i in range(len(input_arr)):
    for j in range(i+1,len(input_arr)):
      if (input_arr[i]+input_arr[j]) == k:
        output.add((min(input_arr[i],input_arr[j]),max(input_arr[i],input_arr[j])))

  return (output)
```

```python
def unique_pairs_sum_2(input_arr,k):
  
  mate= set()
  output= set()
  for i in range(len(input_arr)):
    diff= k - input_arr[i]
    if diff not in mate:
      mate.add(input_arr[i])
    else:
      output.add((min(input_arr[i],diff),max(input_arr[i],diff)))
  
  return (output)
```

**1.12 Find missing element: Consider an array of integers. A second array is formed by shuffling the elements of the first array and deleting a random element. Given the two arrays, find which element is missing in the second array.**

```python
def find_missing_element(input_arr1,input_arr2):
  
  if len(input_arr1) > len(input_arr2):
    a1,a2= sorted(input_arr1), sorted(input_arr2)
  else:
    a1,a2= sorted(input_arr2), sorted(input_arr1)

  for i in range(len(a1)):
    if a1[i] != a2[i]:
      return (a1[i])
```
